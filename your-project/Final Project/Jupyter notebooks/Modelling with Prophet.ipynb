# Modeling economic estimators(Yield&CCI) by Month resampling

```python
yield_cci = pd.read_csv('/Users/vladimirautier/Documents/GitHub/Week9/Project-Week-9-Final-Project/your-project/Final Project/Modified:Cleaned Data/yield_curve_cci_mre.csv')
```

```python
yield_cci.head()
```

```python
yield_cci.dtypes
```

```python
yield_cci.drop(['10 Year Treasury Yield','3 Month Treasury Yield','3 Month Treasury Yield (Bond Equivalent Basis)','Year'],axis=1,inplace=True)
```

```python
yield_cci['Date'] =  pd.to_datetime(yield_cci['Date'])
```

```python
yield_cci.set_index('Date', inplace=True)
```

```python
#Splitting the data

y1 = yield_cci.NBER_Rec
X1= yield_cci.drop('NBER_Rec', axis=1)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
train_X1, test_X1, train_y1, test_y1 = train_test_split(X1, y1, test_size=0.2)
```

```python
#RANDOM FOREST ALGO
```

```python
#Applying the random forest algorithm

from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier()
```

```python
model_rf.fit(train_X1,train_y1)
```

```python
rf_predictions = model_rf.predict(test_X1)
```

```python
rf_probs = model_rf.predict_proba(test_X1)[:, 1]
```

```python
#AUC-SCORE
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_value = roc_auc_score(test_y1, rf_probs)
```

```python
roc_value
```

```python
#R-SQUARED
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = model_rf.predict(train_X1)
```

```python
r2_score(train_y1, y_pred)
```

```python
#CONFUSION MATRIX & ACCURACY
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score 
```

```python
y_pred_CM = model_rf.predict(test_X1)

print(confusion_matrix(test_y1, y_pred_CM))
print(accuracy_score(test_y1, y_pred_CM))
```

```python
#CROSS-VALIDATION
```

```python
from sklearn.model_selection import cross_validate
```

```python
cv_results = cross_validate(model_rf, X1, y1, cv=3)
```

```python
sorted(cv_results.keys())
```

```python
cv_results['test_score']
```
