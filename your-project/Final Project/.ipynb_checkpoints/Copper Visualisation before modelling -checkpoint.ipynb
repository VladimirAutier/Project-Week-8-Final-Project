# Copper price

```python
copper = pd.read_csv('/Users/vladimirautier/Documents/GitHub/Week9/Final Project/copper_clean.csv')
```

```python
copper.columns
```

```python
#Dropping unecessary column
copper.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
# Need to change object type to datetime
copper.dtypes
```

```python
copper['Date'] =  pd.to_datetime(copper['Date'])
```

```python
copper.dtypes
```

```python
copper.head()
```

```python
#renaming columns
copper.columns = ['Time','Value of Copper in Dollars per Tonne'] 
copper.head()
```

```python
# setting the Year column as an index
copper_new = copper.set_index('Time')
copper_new.head()
```

```python
copper_new.index
'''Now that our DataFrame’s index is a DatetimeIndex, we can use all of pandas’ powerful time-based indexing to wrangle and analyze our data, as we shall see in the following sections.'''
```

```python
#Adding new columns to help us later
copper_new['Year'] = copper_new.index.year
copper_new['Month'] = copper_new.index.month
copper_new['Weekday Name'] = copper_new.index.weekday_name
```

```python
copper_new.head()
```

```python
# Setting the figure size for seaborn plots
sns.set(rc={'figure.figsize':(11, 4)})
```

```python
copper_new['Value of Copper in Dollars per Tonne'].plot(linewidth=0.5);
```

```python
cols_plot = ['Value of Copper in Dollars per Tonne', 'Year', 'Month']
axes = copper_new[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('Copper in Dollars per Tonne')
```

```python
#Let’s plot the time series in a single year to investigate further
ax = copper_new.loc['2017', 'Value of Copper in Dollars per Tonne'].plot()
ax.set_ylabel('Value of Copper in Dollars per Tonne');
```

```python
#Further into detail
ax = copper_new.loc['2017-01':'2017-02', 'Value of Copper in Dollars per Tonne'].plot(marker='o', linestyle='-')
ax.set_ylabel('Value of Copper in Dollars per Tonne');
```

```python
#Seasonality
sns.boxplot(data=copper_new, x='Month', y='Value of Copper in Dollars per Tonne');
```

```python
#Frequency
pd.date_range('1980-03-10', '2018-03-15', freq='Q')
```

```python
copper_new.dtypes
```

```python
#copper_new.columns

data_columns = ['Value of Copper in Dollars per Tonne', 'Year', 'Month']
```

```python
copper_365d = copper_new[data_columns].rolling(window=365, center=True, min_periods=360).mean()
```

```python
fig, ax = plt.subplots()

ax.plot(copper_365d['Value of Copper in Dollars per Tonne'], color='0.2', linewidth=3,
label='Trend (365-d Rolling Mean)')
# Set x-ticks to yearly interval and add legend and labels
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Price')
ax.set_title('Value of Copper in Dollars per Tonne');
```
