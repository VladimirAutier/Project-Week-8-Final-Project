# Plotting the same model but with less years to see if the prediction follow the same trend

```python
pd.set_option('display.max_rows', None)
```

```python
df['Year'] = df.index.year
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Using a shorter DF and apply the prophet model. DF1 is the new dataframe
df1 = df.where(df['Year']<2010) 
```

```python
df1.tail()
```

```python
#df.loc['1959': '2000']
```

```python
# Getting rid of NAN values in order to have the dataframe including all values up to 2010.
df1 = df1.dropna()
```

```python
#Applying prophet

model = Prophet()
model_df1 = model.fit(df1)
```

```python
#Forecasting

future_data_1 = model_df1.make_future_dataframe(periods=126, freq = 'm')
```

```python
#Predicting

forecast_data_1 = model_df1.predict(future_data_1)
```

```python
#Predictors & confidence interval

forecast_data_1[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#Change point in the DF = 25
model_df1.changepoints
```

```python
model.plot(forecast_data_1)
```

```python
fig1 = model.plot(forecast_data_1)
a = add_changepoints_to_plot(fig1.gca(), model_df1, forecast_data_1)
```

```python
deltas = model.params['delta'].mean(0)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111)
ax.bar(range(len(deltas)), deltas)
ax.grid(True, which='major', c='gray', ls='-', lw=1, alpha=0.2)
ax.set_ylabel('Rate change')
ax.set_xlabel('Potential changepoint')
fig.tight_layout()
```
